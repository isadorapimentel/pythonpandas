# Selecionando apenas os imóveis residenciais

```python
#importando o pacote pandas 
import pandas as pd
```

```python
#abrindo a base de dados, definindo o separador e inserindo-a na variável dados
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados
```

```python
dados["Tipo"]
```

```python
# retirando valores repetidos da coluna tipo
list(dados['Tipo'].drop_duplicates())
```

```python
#criando uma lista dos tipos que são residenciais e excluindo os demais
residencial = ['Quitinete', 
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
```

```python
#verificando as posições nas quais existem tipos de imóveis residenciais
dados['Tipo'].isin(residencial).head(10)
```

```python
#criando uma seleção apenas com imóveis residenciais a partir da lista criada anteriormento, por meio da função isin
selecao = dados['Tipo'].isin(residencial)
```

```python
# inserindo em uma variável apenas os imóvel selecionados acima como residenciais na variável 'dados_residencia'
dados_residencial = dados[selecao]
```

```python
# verificando quantos imóveis residenciais existem com a função shape
dados_residencial.shape[0] 
```

```python
# reorganizando o indice dos imóveis residenciais
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

# exportando base de dados 

```python
#salvando a base de dados com a selação dos imóveis residenciais
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep = ";", index = False)
```
